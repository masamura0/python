```python
from matplotlib import pyplot
import numpy as np
```

```python
# 折れ線グラフ
values = np.array([1, 2, 3, 4, 5])    # X軸の値
height = np.array([100, 50, 300, 250, 500])    # Y軸の値
pyplot.plot(values, height)    # 値を指定
pyplot.show()  # グラフを描画
```

```python
# 棒グラフ
values = np.array([1, 2, 3, 4, 5])
height = np.array([100, 50, 300, 250, 500])
pyplot.bar(values, height)
pyplot.show()
```

```python
# 円グラフ
values = np.array([0.1, 0.05, 0.30, 0.25, 0.5])
names = ["A", "B", "C", "D", "E"]
pyplot.pie(values,autopct="%d%%", labels = names)
pyplot.axis("equal")
pyplot.show()
```

```python
# ヒストグラム
values = np.random.normal(50, 10, 1000)
pyplot.hist(values)
pyplot.show()
```

```python
# 分布図
x = np.linspace(50,10,100)
y = np.random.randn(100)
pyplot.scatter(x,y)
pyplot.show()
```
